```python
def run_scored(batsman_name):
    # Filter for the given batsman
    mask = delivery[delivery['batsman'] == batsman_name]
    
    # Check if the filtered DataFrame is empty
    if mask.empty:
        return f"No data found for batsman: {batsman_name}"
    
    # Group by 'bowling_team' and calculate the sum of 'batsman_runs'
    grouped = mask.groupby('bowling_team')['batsman_runs'].sum()
    
    # Check if the grouped DataFrame is empty
    if grouped.empty:
        return f"No runs data available for batsman: {batsman_name}"
    
    # Return the team against which the batsman scored the most runs
    return grouped.sort_values(ascending=False).index[0]
```

```python
run_scored('V Kohli')
```

**merge()function**

```python
# it is used to merge the two dataFrame based on common column
# reset_index() is used to convert the series into dataFrame
```

```python
# Find the orange cap holder of each year(season) and this is find with the help of merge() function
new = delivery.merge(
    right=df,             # Right DataFrame
    left_on='match_id',   # Key from the left DataFrame
    right_on='id',        # Key from the right DataFrame
    how='inner'           # Type of merge
)

# groupby and after drop the duplicate row and keep = first
result =(
        new.groupby(['season', 'batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()
        .drop_duplicates(subset = 'season', keep = 'first').sort_values(by = 'season', ascending=True)
)
# print(result) -> This print the series 
result
```

***Team Performance Analysis***

```python
#Plot the top 5 team whose played most of the matches in ipl till 2017

# Calculate total matches played by each team
total_matches = (df['team1'].value_counts() + df['team2'].value_counts()).sort_values(ascending=False).head(5)

# Convert to DataFrame
total_matches_df = total_matches.reset_index()
total_matches_df.columns = ['Team', 'Matches Played']

# Set the figure size
plt.figure(figsize=(10, 5))
ax = sns.barplot(data=total_matches_df, x='Team', y='Matches Played', hue='Team', palette='dark', legend=False)

# Add labels to bars
for bars in ax.containers:
    ax.bar_label(bars)

# Customize the plot
plt.xlabel("Team", fontsize=12)
plt.ylabel("Total Matches Played", fontsize=12)
plt.title("Top 5 Teams by Matches Played", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Winning Percentage for Each Team -> Top 5 teams
team_wins = df['winner'].value_counts()
team_matches = df['team1'].value_counts() + df['team2'].value_counts()
win_percentage = (team_wins / team_matches * 100).dropna().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
ax = win_percentage.plot(kind='bar', color='teal', title='Winning Percentage of Teams')

# Add labels on top of each bar
ax.bar_label(ax.containers[0], fmt='%.1f%%', padding=2, fontsize=7)

plt.show()
```

# Final Conclusion of IPL Analysis

**1. Team Performance Trends**
   * Certain teams, like Mumbai Indians (MI) and Chennai Super Kings (CSK), have been the most successful based on title wins and consistency in playoffs.
   * Some teams perform exceptionally well in league stages but struggle in knockouts.


**2. Impact of Toss and Venue**
   * Toss plays a crucial role in match outcomes, especially in venues where chasing is easier.
   * Teams winning the toss often choose to chase, as batting second has historically provided an advantage.


**3. Player Performance Insights**
   * Players like Virat Kohli, Rohit Sharma, and David Warner have consistently been among the top run scorers.
   * Bowlers like Lasith Malinga, Jasprit Bumrah, and Yuzvendra Chahal have proven crucial in T20 success.
   * All-rounders like Hardik Pandya, Ravindra Jadeja, and Andre Russell significantly impact match results.


**4. Batting vs. Bowling Dominance**
   * In earlier seasons, batting-heavy teams dominated, but in recent seasons, balanced teams with strong bowling attacks have been more successful.
   * Death-over specialists and powerplay bowlers play a key role in restricting opponents.

**5. Emerging Player Trends**
   * IPL continues to be a great platform for young talents, with players like Shubman Gill, Ruturaj Gaikwad, and Umran Malik rising to prominence.
   * Franchises invest heavily in domestic talent, ensuring a steady influx of fresh players.


**6. Win Patterns and Strategies**
   * Teams that excel in powerplays and death overs tend to have higher winning percentages.
   * Partnerships and middle-over stability are crucial for setting up big totals or successful chases.

     
**7. Economic and Fan Engagement Impact**
   * The IPL has grown into a multi-billion-dollar league, attracting global sponsorships and high viewership.
   * Social media engagement and fantasy leagues have increased fan participation and interest in cricket.
